```python
cities = df['Reference Area'].tolist()
```

```python
import random
def ploting(x,y,title):
    colors = ['red','blue','green','orange','purple','cyan','magenta','yellow','lime','gold']
    sorted_indices = np.argsort(x)
    sorted_cities = [cities[i] for i in sorted_indices]
    sorted_crime_data = [x[i] for i in sorted_indices]
    plt.figure(figsize=(12,8))
    t = title.split(' ')
    a = 'Crime Ranking for ' + title + ' Offences In Different Cities'
    plt.title(a)
    if y == 1:
        plt.bar(sorted_cities[::-1], sorted_crime_data[::-1], color = colors[random.randint(0,9)])
        plt.xlabel('Cities')
        plt.ylabel(t[0] + '_Offences')
    else:
        plt.barh(sorted_cities, sorted_crime_data, color = colors[random.randint(0,5)])
        plt.ylabel('Cities')
        plt.xlabel(t[0]+'_Offences')
    plt.xticks(rotation = 90)
    plt.tight_layout()
    plt.show()
```

```python
ploting(df['All Crimes'].tolist(),0,'Crime')
ploting(df['All Group 1: Non-sexual crimes of violence'].tolist(),0,'Non_Sexual')
ploting(df['All Group 2: Sexual crimes'].tolist(),1,'Sexual')
ploting(df['All Group 3: Crimes of dishonesty'].tolist(),1,'Dishonesty')
ploting(df['All Group 4: Damage and reckless behaviour'].tolist(),1,'Reckless_Behaviour')
ploting(df['All Group 5: Crimes against society'].tolist(),1,'Crime_Against_Society')
ploting(df['All Group 6: Antisocial offences'].tolist(),1,'AntiSocial')
ploting(df['All Group 7: Miscellaneous offences'].tolist(),1,'Miscellaneous')
ploting(df['All Group 8: Road traffic offences'].tolist(),1,'Road_Traffic')
ploting(df['All Offences'].tolist(),1,'All')
```

```python
crime_data_all_offences = df['All Offences'].tolist()
```

```python
# Sort cities and crime data in descending order
sorted_indices_descending = np.argsort(crime_data_all_offences)[::-1]
sorted_cities_descending = [cities[i] for i in sorted_indices_descending]
sorted_crime_data_descending = [crime_data_all_offences[i] for i in sorted_indices_descending]

# Top ten highest ranking cities
top_ten_highest_ranking_cities = sorted_cities_descending[:10]
top_ten_highest_ranking_crime_data = sorted_crime_data_descending[:10]

# Top ten lowest ranking cities
top_ten_lowest_ranking_cities = sorted_cities_descending[-10:]
top_ten_lowest_ranking_crime_data = sorted_crime_data_descending[-10:]

# Plot top ten highest ranking cities
plt.figure(figsize=(12, 6))
plt.bar(top_ten_highest_ranking_cities, top_ten_highest_ranking_crime_data, color='red')
plt.xlabel('City')
plt.ylabel('Number of Crimes')
plt.title('Top Ten Highest Ranking Cities for "All Offences"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot top ten lowest ranking cities
plt.figure(figsize=(12, 6))
plt.bar(top_ten_lowest_ranking_cities, top_ten_lowest_ranking_crime_data, color='green')
plt.xlabel('City')
plt.ylabel('Number of Crimes')
plt.title('Top Ten Lowest Ranking Cities for "All Offences"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Analysis:
Top Ten Highest Ranking Cities:

The analysis identified the top ten cities with the highest number of reported crimes under the category "All Offences".
Glasgow City tops the list with a significantly higher number of reported crimes compared to other cities.
Other cities in the top ten list include City of Edinburgh, North Lanarkshire, Fife, and Aberdeen City, among others.
The bar chart visualizes the crime data for each city, highlighting Glasgow City as the city with the highest crime rate.
Top Ten Lowest Ranking Cities:

Conversely, the analysis also identified the top ten cities with the lowest number of reported crimes under the category "All Offences".
These cities exhibit relatively lower crime rates compared to the top-ranking cities.
Orkney Islands, Na h-Eileanan Siar, and Shetland Islands are among the cities with the lowest crime rates.
The bar chart illustrates the crime data for each city, emphasizing the comparatively lower crime rates in these regions.

While some cities experience higher crime rates, others demonstrate relatively lower levels of criminal activity.
